This notebook will show you how to use *pymaid.plot* to generate 2d and 3d plots of your neurons of interest.

```python
from pymaid.pymaid import CatmaidInstance
from pymaid.plot import plot2d, plot3d

#Create Catmaid instance
rm = CatmaidInstance( 'https://www.your.catmaid-server.org' , 'http_user' , 'http_password', 'token' )
```

First, let's try to generate a 2d plot. This currently features only frontal (x/y) views. Feel free to check out help(plot2d) and help(plot3d) for more details.

```python
fig, ax = plot2d( skids = [ 2333007 ], remote_instance = rm, connectors = True )
```

*plot2d()* generates a matplotlib object, we have to call *matplotlib.pyplot.show()* in order to render it. If you like, you can adjust the plot by accessing properties of *fig* and *ax*.

```python
import matplotlib.pyplot as plt
plt.show()
```

Alternatively use **plt.savefig(filename.svg)**to save a editable vector graphic file. Next, let's try rendering a different neuron in 3D using [plotly](https://plot.ly/). We can even include 3D volumes from your CATMAID project.

```python
from plotly import offline as pyoff

#Need to initialize plotly for interactive rendering with Jupyter notebooks
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
fig = plot3d(
                  skids = [ 1191261 ], 
                  remote_instance = rm, 
                  connectors = False, 
                  fig_autosize = True,
                  volumes = { 'v13.LH_R':  ( 200,200,255 ),
                              'v13.LH_L': ( 200,200,255 ) } 
                 )
```

*plot3d* returns a dictionary that we can pass to plotly to generate a 3d plot. Here, we use **plotly.offline.iplot(fig)** which is used in Jupyter notebooks. If you want to plot3d to file, use **plotly.offline.plot(fig, filename.html)**. Please note that the plot (e.g. line width, etc) may look differently depending on what browser you are using. Generally, layout is optimised for Google Chrome.

```python
pyoff.iplot( fig )
```
